```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text, 'html.parser')
table = soup.find('table', class_='TABLE_G')
print(table)
```

```python
#將每個指標及數據都存成列表，轉成一個dict，再存成一個大list(後續存成一個大dict用)
tags=table.find_all('td') 

n=0
datalist=[]

while n<350: #tags共350個<td>標籤
    datelist=tags[n+2:n+33:3] #取出每個指標的日期
    valuelist=tags[n+3:n+34:3] #取出每個日期的量測值
    #整理成列表
    date=[]
    value=[]
    for dates in datelist:
        date.append(dates.string)
    for values in valuelist:
        value.append(values.string)

    dic=dict(zip(date,value))#可用dict(zip(list1,list2))將兩列表變成一個字典
    datalist.append(dic) #將不同指標字典累加

    #若想分別印出字典:   用 locals()可循環命名變數  
    #     locals()['s{}'.format(i)]=dict(zip(date,value))                                                   
    #     if i <=10:  設定迴圈增加i值
    #         print('s{}='.format(i),dict(zip(date,value))) 
    
    n+=35
    
#print(datalist)必要時可印出參考

#存成適合的資料形式
#將所有指標 &資料 存成一個大dict 
d=dict(zip(['SO2','CO','O3','PM10','NOx','NO','NO2','THC','NMHC','CH4'],datalist))
print(d)
```
