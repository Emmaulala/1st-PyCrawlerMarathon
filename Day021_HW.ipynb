## 1. 取出今天所有的新聞

```python
import datetime
from selenium import webdriver 
from bs4 import BeautifulSoup
import time
import re
```

```python
#定義函數迴圈scroll找資料
browser = webdriver.Chrome(executable_path='../chromedriver') #用google chrome專用driver啟動瀏覽器(讓它代為做操作)
browser.get("https://www.ettoday.net/news/news-list.htm")

today=str(datetime.date.today()).replace('-','/') #取得今天日期
pattern=re.compile(today+'.+') #將正則表達式編譯成pattern

#scroll找資料，若沒找到繼續回來scroll
def ScrollPage(n):
    time.sleep(1) 
    browser.execute_script("window.scrollTo(0,100000);") #設定極大座標，讓網頁不停想滑到最底部
    html_source = browser.page_source #取得資料
    soup = BeautifulSoup(html_source, "html5lib") #解析html資料
    lastnewsdate=soup.find(class_="part_list_2").find_all('h3')[-1].text #最後一則新聞的日期
    
    if pattern.match(lastnewsdate): #若最後一則新聞日期還是當天日期
        ScrollPage(n) #持續scroll
    else:
        for d in soup.find(class_="part_list_2").find_all('h3'):
            if pattern.match(d.find(class_="date").text): #若新聞日期匹配當天日期則印出
                print(d.find(class_="date").text, d.find_all('a')[-1].text)
            else:
                continue
        
ScrollPage(1) 
```

## 2. 取出現在時間兩小時內的新聞

```python
browser.execute_script("window.scrollTo(0,100000);") #設定極大座標，讓網頁不停想滑到最底部
html_source = browser.page_source #取得資料
soup = BeautifulSoup(html_source, "html5lib") #解析html資料

#統一將時間轉換為 str或 datetime.datetime資料格式做比較
timenow=datetime.datetime.now() #取得現在時間
time2hour=timenow-datetime.timedelta(hours=2) #利用datetime.timedelta()獲取兩小時前的時間
time2hour=str(time2hour).replace('-','/')

for d in soup.find(class_="part_list_2").find_all('h3'):
    newstime=d.find(class_="date").text #找到日期時間並轉成統一格式
    if newstime > time2hour: #若新聞時間比兩小時前時間還後面
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
    else:
        continue
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
#取得三天前日期
threeday=str(datetime.date.today()-datetime.timedelta(days=3)).replace('-','/')
year=threeday.split('/')[0]
month=threeday.split('/')[1]
day=threeday.split('/')[2]
```

```python
#讓driver驅動js選擇到三天前日期按查詢
from selenium.webdriver.support.ui import Select
browser = webdriver.Chrome(executable_path='../chromedriver')#注意chromedriver的path
browser.get("https://www.ettoday.net/news/news-list.htm")

selectyear = Select(browser.find_element_by_id("selY"))
selectyear.select_by_value(year)#可從開發人員工具Elements中找出
selectmonth = Select(browser.find_element_by_id("selM"))
selectmonth.select_by_value(month)
selectday = Select(browser.find_element_by_id("selD"))
selectday.select_by_value(day)

browser.find_element_by_id('button').click()

#scroll直到滑到三點前的新聞
n=1
while n>0:
    time.sleep(1) 
    browser.execute_script("window.scrollTo(0,100000);") 
    html_source = browser.page_source #取得資料
    soup = BeautifulSoup(html_source, "html5lib") #解析html資料
    lastnewsdate=soup.find(class_="part_list_2").find_all('h3')[-1].text #最後一則新聞的日期

    if lastnewsdate>'2019/12/30 15:00': #若最後一則新聞在下午三點之後，持續scroll
        n=1
    else:
        for d in soup.find(class_="part_list_2").find_all('h3'):
            if '2019/12/30 15:00'<d.find(class_="date").text<'2019/12/30 17:00': #若新聞日期在指定區間內
                print(d.find(class_="date").text, d.find_all('a')[-1].text)
            else:
                continue
        n=0

```
