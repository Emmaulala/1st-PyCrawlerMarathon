### ① 取出今天所有的發文

```python
import requests

url='https://www.ettoday.net/news/news-list.htm'
r=requests.get(url)
data=r.text #type:str
```

```python
from bs4 import BeautifulSoup

res=BeautifulSoup(data,'html.parser')
print(res)
```

```python
#方法一:
data=res.find('div',class_='part_list_2').find_all('h3')

for d in data:
    newstime=d.find(class_='date') #注意此處 .find()和 .find_all()差別
    newssort=d.find('em')
    newstitle=d.find('a')
    print(newstime.text,'[{}]'.format(newssort.string),newstitle.text)
    #注意 .string和 .text差別，參考https://zhuanlan.zhihu.com/p/30911642
```

```python
#方法二:
import pandas as pd 
time=[]
sort=[]
title=[]
for d in data:
    newstime=d.find(class_='date') #注意此處 .find()和 .find_all()差別
    time.append(newstime.string)
    newssort=d.find('em')
    sort.append(newssort.string)
    newstitle=d.find('a')
    title.append(newstitle.text)

#建立dataframe
dataframe=pd.DataFrame({
    'Released time':time,
    'Sort':sort,
    'News title':title})
pd.set_option('display.max_rows', None) 
print(dataframe)
# 若結果被省略，可用pd.set_option('display.max_rows', None)顯示所有列數 和pd.set_option('display.max_columns', None)顯示所有欄數
```

```python
#方法三:
dataresult=pd.DataFrame(data,columns=['news time','news sort','news title'])
dataresult 
#此處建立的dataframe是以原本解析後data直接建立，裡面含有標籤，會影響後續groupby等操作
#建立dataframe即使列表有標籤還是會只建立文字部分，小心後續groupby等操作，無法支持含有標籤的分類
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
#方法一:
data=res.find('div',class_='part_list_2').find_all('h3') #len(data)共有100個結果

def sortdata(n):
    i=0
    for d in data:
        if n in (d.find('em')['class'])[-1]: #若n字串在標籤em的屬性class中(須指定list中的最後一項[-1]，因屬性class的值有tag和 c_news...等兩樣，結果為列表)
            newssort=d.find('em')
            newstime=d.find(class_='date') 
            newstitle=d.find('a')
            print(newstime.text,'[{}]'.format(newssort.string),newstitle.text)
            i+=1
    print('total:{} news'.format(i)) 
    

#呼叫函式即可將每一類新聞取出
sortdata('news')
sortdata('sport')
sortdata('star')
```

```python
#方法二:

def sortdatalist(n):
    newstimelist=[]
    newstitlelist=[]
    for d in data:
        if n in (d.find('em')['class'])[-1]: #必須指定list中的最後一項[-1]，因屬性class的值有tag和 c_news...等兩樣，結果為列表
            newssort=d.find('em').string
            newstime=d.find(class_='date') 
            newstimelist.append(newstime.string)
            newstitle=d.find('a')
            newstitlelist.append(newstitle.text)
            #print(newstime.text,'[{}]'.format(newssort.string),newstitle.text)
        length=len(newstitlelist)
        sortDataframe=pd.DataFrame({'time':newstimelist,'title':newstitlelist})
    print(newssort,'{} news'.format(length))
    print(sortDataframe)

sortdatalist('star') #呼叫函式
```

### ③ 哪一個類別的文章最多

```python
time=[]
sort=[]
title=[]
for d in data:
    newstime=d.find(class_='date') #注意此處 .find()和 .find_all()差別
    time.append(newstime.string)#需用 .string 或 .text去除標籤，否則影響 groupby等操作
    newssort=d.find('em')
    sort.append(newssort.string)
    newstitle=d.find('a')
    title.append(newstitle.text)
#建立dataframe
dataframe=pd.DataFrame({
    'Released time':time,
    'Sort':sort,
    'News title':title})
print(dataframe) 

#用 pandas內建函式 groupby分組做數據挖掘
datagroup=dataframe.groupby('Sort') 

print(datagroup.size())#用.size()看每一組大小
datagroup.groups #.groups看分組結果
datagroup.get_group('影劇') #.get_group取出某一組
```
