## 作業目標

* 目前程式中的 proxy_ips 是手動輸入的，請根據 https://free-proxy-list.net/ 寫一個可自動化抓取可用 Proxy 的 proxy_ips。




```python
from bs4 import BeautifulSoup
import requests
import random

#送出請求解析網站

proxy_ips = []
url='https://free-proxy-list.net/'
res=requests.get(url)
soup=BeautifulSoup(res.text,'html.parser')

#找到標籤取出ip和port資訊

tags=soup.find('table',id='proxylisttable').find_all('td')
datalist=[]
for tag in tags:
    ip=tag.text
    datalist.append(ip)   
iplist=datalist[0::8]
portlist=datalist[1::8]

#組成完整ip

for n in range(len(iplist)):
    ip=iplist[n]
    port=portlist[n]
    proxy_ip=ip+':'+port
    proxy_ips.append(proxy_ip)

print(proxy_ips)
```

```python
for i in range(10):
    ip = random.choice(proxy_ips) #隨機切換ip增強爬蟲
    print('Use', ip)
    try: #試用此ip是否可爬取到資料
        resp = requests.get('http://ip.filefab.com/index.php',
                        proxies={'http': ip, 'https': ip}, timeout=10) # timeout設置等待多久超時，解決某些網站過慢影響 (此處最多跑10秒自動停止)
        soup = BeautifulSoup(resp.text, 'html5lib')
        print(soup.find('h1', id='ipd').text.strip()) 
    except:
        print('Fail')
```
