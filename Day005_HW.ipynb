```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
import requests
r=requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
r.text #觀察資料為json格式的字串
```

```python
#匯入解析資料所需的模組
import json
response=r.text 
data=json.loads(response)
length=len(data)
print('此API回傳{}筆資料'.format(length))#回傳的資料數統計
print('每筆資料包含以下欄位:')
dict.keys(data[0])#獲取第一個字典的所有key值
```
